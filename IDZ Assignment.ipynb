```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('IDZ Digital data analyst.xls')
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.suptitle("Gender Distribution in Dataset", size=20)
sns.countplot(x=df.Gender)
```

* Total number of Male are more in the data as compared to Female.

```python
plt.figure(figsize=(10,5))
plt.suptitle("Total Number of Married and Unmarried peoples in dataset", size=20)
sns.countplot(x=df.MaritalStatus)
```

* Total number of Single peoples are more in the data as compared to married peoples.

```python
plt.figure(figsize=(20,8))
plt.suptitle("Occupation distribution in dataset ", size=20)
sns.countplot(y=df.Occupation)
```

* As compare to other occupation highest number of people are work as Insurer in given data while lowest number of people are scientist

```python
df.groupby('Occupation').mean().sort_values('Salary',ascending=True).head(10)['Salary']
```

```python
plt.figure(figsize=(20,8))
plt.barh(df['Occupation'],df['Salary'].sort_values())
plt.show()
```

* Mechanic,Aeroplane Pilot,Accountant,Firefighter,Police Officer has highest salary and Photographer,Cook,Florist,Jeweller,     Manager has lowest salary

```python
plt.figure(figsize=(10,7))
plt.title('Age Distribution of Data')
sns.countplot(df.Age)
plt.show()
```

* Maximum number of people has age group 55

```python
plt.figure(figsize=(10,7))
plt.title('Age Distribution of Data')
sns.histplot(df.Age)
plt.show()
```

```python
print('The max Salary:',df.Age.max())
print('The min Salary:',df.Age.min())
```

* From the above charts , it is clear that there all the ages are more or less equally distruibuted across the data set.

```python
plt.figure(figsize=(10, 8))
plt.title("Salary distribution in dataset", size=20)
plt.xlabel('Salary Range')
sns.histplot(df['Salary'])
```

```python
print('The max Salary:',df.Salary.max())
print('The min Salary:',df.Salary.min())
```

```python
plt.figure(figsize=(20,8))
plt.suptitle("Salary diffrence in Male and Female", size=20)
sns.histplot(x='Salary',hue='Gender',data=df,multiple='dodge')
```

* From the above charts , it is clear that , there exist gap in male & female Salary 

```python
plt.figure(figsize=(10,8))
df.groupby('Gender')['Salary'].mean().plot(kind='bar')
plt.title("Avg Salary Diffrence between Genders ")
plt.show()
```

* From the above charts Males are earning more as compair to female

```python
plt.figure(figsize=(20,8))
df.groupby('Occupation')['Salary'].mean().plot(kind='bar')
plt.title("Avg Salary of Diffrence between Occupations ")
plt.show()
```

```python
fig1,ax1 = plt.subplots()
colors = ['gold','lightskyblue']
ax1.pie(Values,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90,colors=colors)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.suptitle("Top most popular job among Male", size=20)
df[df.Gender=='Male'].Occupation.value_counts().iloc[0:11].plot(kind='bar')
plt.xlabel('Occupation Title')
```

* Insure is the most popular job amoung males

```python
plt.figure(figsize=(12, 8))
plt.suptitle("Top most popular job among Female", size=20)
df[df.Gender=='Female'].Occupation.value_counts().iloc[0:11].plot(kind='bar')
plt.xlabel('Occupation Title')
```

* Social worker and Lawer are top most popular jobs among Female.

```python
plt.figure(figsize=(15, 8))
plt.suptitle("Top 10 avg high paying Occupation in dataset", size=10)
df.groupby('Occupation').mean().sort_values('Salary',ascending=False).head(10)['Salary'].plot(kind='bar')
plt.xlabel('Occupation')
```

* This graph showing Top 10 avg highest paying Occupation in dataset.

```python
plt.figure(figsize=(15, 8))
plt.suptitle("Top 10 avg low paying Occupation in dataset", size=10)
df.groupby('Occupation').mean().sort_values('Salary',ascending=True).head(10)['Salary'].plot(kind='bar')
plt.xlabel('Occupation')
```

* This graph showing Top 10 avg low paying Occupation in dataset.
